```
import io
import os
import pandas as pd

# Imports the Google Cloud client library
from google.cloud import vision
from google.cloud.vision import types

os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/content/drive/My Drive/CS298-bab074d5fcb2.json"

# Instantiates a client
client = vision.ImageAnnotatorClient()

df = pd.read_csv('/content/drive/My Drive/dataSet3.csv')
df.set_index('image', inplace=True)

directory ='/content/drive/My Drive/dataset_298/english'

def detect_context(path, image_name):
  # Loads the image into memory
  with io.open(path, 'rb') as image_file:
    content = image_file.read()
  image = types.Image(content=content)
  # Performs label detection on the image file
  response = client.label_detection(image=image)
  labels = response.label_annotations
  label_str = ""
  for label in labels:
    #print(label.description)
    label_str = label_str + label.description + "\n"
  df.loc[image_name,'context'] = label_str
  print(label_str)
    
for filename in os.listdir(directory):
#  filename = '8.png' #remove this
  if(filename != ".DS_Store"):
    file_name = os.path.abspath('/content/drive/My Drive/dataset_298/english/'+filename)
    detect_context(file_name, filename)
    print(filename)
#    break # remove this
df.to_csv('/content/drive/My Drive/dataSet4.csv')

```
